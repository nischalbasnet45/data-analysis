```python
# Outlier Detection - IQR Method for Age
plt.figure(figsize=(8, 6))

# Calculate IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Box plot with outliers highlighted
sns.boxplot(x='Age', data=df)
sns.scatterplot(x='Age', y=0, data=outliers, color='red', marker='o', label='Outliers')

plt.title('Box Plot of Age with Outlier Detection')
plt.legend()
plt.show()
```

```python
for i, feature in enumerate(df.columns):
    if df[feature].dtype == 'int64' or df[feature].dtype == 'float64':
        # Calculate IQR
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]

        # Plot the outlier detection graph
        sns.boxplot(x=feature, data=df)
        sns.scatterplot(x=feature, y=0, data=outliers, color='red', marker='o', label='Outliers')
        plt.title(f'Box Plot of {feature} with Outlier Detection')
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load your dataset
file_name = "WFH_WFO_dataset.csv"
df = pd.read_csv(file_name)

# Drop non-numeric columns or encode them if needed
X = df.drop(['Target', 'Name'], axis=1)

# Encode categorical variables if needed
X_encoded = pd.get_dummies(X, drop_first=True)

# Extract the prediction label
y = df['Target']

# Standardize the data before applying PCA
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X_encoded)

# Apply PCA
pca = PCA(n_components=2)  # You can adjust the number of components
X_pca = pca.fit_transform(X_standardized)

# Create a DataFrame with the PCA components and the prediction label
df_pca = pd.DataFrame(data={'PCA1': X_pca[:, 0], 'PCA2': X_pca[:, 1], 'Target': y})

# Visualize the data in 2D using PCA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Target', data=df_pca, palette='viridis')
plt.title('2D Visualization using PCA')
plt.show()
```
